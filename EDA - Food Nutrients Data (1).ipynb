```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "Food And Nutrient Database For Dietary Studies/2021-2023 FNDDS At A Glance - FNDDS Nutrient Values.xlsx"
xls = pd.ExcelFile(file_path)
df = pd.read_excel(xls, sheet_name="FNDDS Nutrient Values", skiprows=1)
df.columns = df.columns.str.strip().str.replace("\n", " ")
print("Dataset Overview:")
print(df.info())
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Energy (kcal)"], bins=30, kde=True)
plt.title("Distribution of Energy (kcal)")
plt.xlabel("Energy (kcal)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Distribution of Macronutrients
fig, axes = plt.subplots(1, 3, figsize=(18,5))
sns.histplot(df["Protein (g)"], bins=30, kde=True, ax=axes[0])
axes[0].set_title("Protein (g) Distribution")
sns.histplot(df["Carbohydrate (g)"], bins=30, kde=True, ax=axes[1])
axes[1].set_title("Carbohydrate (g) Distribution")
sns.histplot(df["Total Fat (g)"], bins=30, kde=True, ax=axes[2])
axes[2].set_title("Total Fat (g) Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,8))
# sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm", annot=False)
sns.heatmap(df.corr(), cmap="coolwarm", annot=False)
plt.title("Correlation Matrix of Nutrients")
plt.show()
```
